```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Conv2D,MaxPooling2D,Flatten,Dense,Dropout
```

```python
(X_train,Y_train),(X_test,Y_test)=mnist.load_data()
X_train=X_train/255.0
X_test=X_test/255.0
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
fig,ax=plt.subplots(ncols=4,figsize=(20,20))
for i in range(4):
  ax[i].imshow(X_train[i],cmap='gray')
plt.show()
```

```python
from tensorflow.keras import optimizers,layers
```

```python
input=Input((28,28,1))
c1=Conv2D(32,(3,3),activation='relu')(input)
c1=MaxPooling2D((2,2))(c1)

c2=Conv2D(64,(3,3),activation='relu')(c1)
c2=MaxPooling2D((2,2))(c2)

c3=Conv2D(128,(3,3),activation='relu')(c2)
c3=MaxPooling2D((2,2))(c3)

flatten=Flatten()(c3)
x=Dense(512,activation='relu')(flatten)
outputs=Dense(1,activation='sigmoid')(x)

model=tf.keras.Model(inputs=[input],outputs=[outputs])
```

```python
model.compile(tf.keras.optimizers.Adam(learning_rate=0.001),loss=tf.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
histroy=model.fit(X_train,Y_train,validation_split=0.2,epochs=10,batch_size=32)
```

```python
plt.plot(histroy.history['accuracy'],label='train_acc')
plt.plot(histroy.history['val_accuracy'],label='val_acc')
plt.legend()
plt.show()
```

```python
test_loss,test_accuracy=model.evaluate(X_test,Y_test)
print("Test Loss:",test_loss)
print("Test Accuracy:",test_accuracy)
```
